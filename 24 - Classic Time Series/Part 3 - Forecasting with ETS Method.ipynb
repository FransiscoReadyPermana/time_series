# Excercise 1: Cheese Production Data

```python
df = pd.read_csv("data/cheese_fred.csv", index_col="date", parse_dates=["date"])
series = df["total_cheese_production"]
X_train, X_test = series[:-25], series[-25:]
df.head()
```

```python
component = additive_decomposition(series, plot=True)
```

```python
series.plot(figsize=(12,5))
```

```python
es = ExponentialSmoothing(X_train, trend="add", seasonal="add", seasonal_periods=12, use_boxcox=True)
model = es.fit(optimized=True)
```

```python
model.predict(start=X_test.index[0], end=X_test.index[-1]).plot(label="prediction", figsize=(12,5))
```

```python
X_train.shape
```

```python
model.predict(len(X_train)-10, len(X_train)+50).plot(label="prediction", figsize=(12,5))
```

```python
from turtle import color

pred = model.predict(len(X_train)-10, len(X_train)+50)

plt.figure(figsize=(15,5))

plt.plot(X_train, label="train", color="blue")
plt.plot(X_test, label="test", color="orange")
plt.plot(pred, label="prediction", color="green")
plt.legend()
```

```python
es = ExponentialSmoothing(X_train, seasonal="add", seasonal_periods=12, use_boxcox=True)
model = es.fit(optimized=True)

pred = model.predict(len(X_train)-10, len(X_train)+50)

plt.figure(figsize=(15,5))

plt.plot(X_train, label="train", color="blue")
plt.plot(X_test, label="test", color="orange")
plt.plot(pred, label="prediction", color="green")
plt.legend()
```
